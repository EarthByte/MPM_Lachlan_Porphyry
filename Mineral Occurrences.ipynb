```python
import warnings
warnings.filterwarnings('ignore')

import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as cx
from ipywidgets import interact
```

### Cu-bearing Mineral Occurrences

```python
mineral_occ = gpd.read_file('./Datasets/Mineral Occurrences/mineral_occurrences.shp')
nsw_boundary = gpd.read_file('./Datasets/Frames/NSW_Boundary.shp')
lachlan_boundary = gpd.read_file('./Datasets/Frames/Lachlan_Boundary.shp')

commodity = 'Cu'

fig, ax = plt.subplots(figsize=(15, 15))
data_minor = mineral_occ[mineral_occ['MINOR_COMM'].str.contains(commodity)==True] # minor commodity
data_major = mineral_occ[mineral_occ['MAJOR_COMM'].str.contains(commodity)==True] # major commodity
data_minor.plot(ax=ax, edgecolor='black', color='yellow', markersize=15, alpha=0.5)
data_major.plot(ax=ax, edgecolor='black', color='orange', markersize=15, alpha=0.5)
nsw_boundary.plot(ax=ax, edgecolor='black', color='none', linewidth=2)
lachlan_boundary.plot(ax=ax, edgecolor='red', color='none', linewidth=2)
cx.add_basemap(ax, crs='EPSG:4283', source=cx.providers.Esri.WorldGrayCanvas)
ax.set_title(commodity+'-bearing Mineral Occurrences')
plt.show()
```

### Mineral deposit classification - NSW classification type

```python
mineralisation_types = data_major.NSW_CLASS.unique()
# mineralisation_types.sort()

print(data_major.NSW_CLASS.value_counts())

@interact(mineralisation_type=mineralisation_types)
def show_map(mineralisation_type):
    fig, ax = plt.subplots(figsize=(15, 15))
    data = data_major[data_major.NSW_CLASS==mineralisation_type]
    data.plot(ax=ax, edgecolor='black', color='yellow', markersize=15, alpha=0.5)
    nsw_boundary.plot(ax=ax, edgecolor='black', color='none', linewidth=2)
    lachlan_boundary.plot(ax=ax, edgecolor='red', color='none', linewidth=2)
    cx.add_basemap(ax, crs='EPSG:4283', source=cx.providers.Esri.WorldGrayCanvas)
    ax.set_title(mineralisation_type)
    plt.show()
```
